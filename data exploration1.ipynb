```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load iris dataset from sklearn
iris = load_iris()

# Convert it into a DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target column (species)
df['species'] = iris.target
df['species'] = df['species'].apply(lambda x: iris.target_names[x])

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# If there were any missing values, you could handle them like this:
# df.fillna(df.mean(), inplace=True)
```

```python
# Pairwise scatter plots for features
plt.figure(figsize=(12, 8))
pd.plotting.scatter_matrix(df, figsize=(12, 8), diagonal='kde', c=df['species'].apply(lambda x: {'setosa': 0, 'versicolor': 1, 'virginica': 2}[x]), cmap='viridis')
plt.show()
```

```python
# Boxplot for each feature by species
df.boxplot(by='species', figsize=(12, 6))
plt.tight_layout()
plt.show()
```

```python
# Histogram for each feature
for column in df.columns[:-1]:
    plt.figure()
    df.groupby('species')[column].plot(kind='hist', alpha=0.5, legend=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
